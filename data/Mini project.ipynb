```python
# =====================================================================
# OOP MINI PROJECT: SMART DEVICE MANAGEMENT SYSTEM
# Course: EL 162 / 234 | Instructor: Dr. Matthew Cobbinah
# =====================================================================

# --- 1. PARENT CLASS: SmartDevice ---
class SmartDevice:
    """Represents a generic smart device with encapsulated core properties."""

    def _init_(self, name: str, device_id: str):
        # Public Attribute
        self.name = name

        # Private Attributes (Encapsulation)
        if not device_id.strip():
            raise ValueError("Device ID cannot be empty.")
        self.__device_id = device_id
        self.__power_status = False  # Defaults to Off (False)

    # Getter and Setter for device_id
    @property
    def device_id(self) -> str:
        return self.__device_id

    # Getter and Setter for power_status
    @property
    def power_status(self) -> bool:
        return self.__power_status

    # Methods
    def turn_on(self):
        self.__power_status = True
        print(f"[{self.name}] has been turned ON.")

    def turn_off(self):
        self.__power_status = False
        print(f"[{self.name}] has been turned OFF.")

    def display_info(self):
        status = "ON" if self.__power_status else "OFF"
        print(f"Device Name: {self.name} | ID: {self.__device_id} | Power: {status}")


# --- 2. CHILD CLASS: TemperatureSensor ---
class TemperatureSensor(SmartDevice):
    """Child class inheriting from SmartDevice to read environment temperature."""

    def _init_(self, name: str, device_id: str, temperature: float = 22.0):
        # Initialize inherited attributes using super()
        super()._init_(name, device_id)
        self.temperature = temperature

    def read_temperature(self):
        if not self.power_status:
            print(f"Cannot read temperature. [{self.name}] is OFF.")
            return
        print(f"[{self.name}] current temperature reading: {self.temperature}°C")


# --- 3. CHILD CLASS: SmartLight ---
class SmartLight(SmartDevice):
    """Child class inheriting from SmartDevice to control brightness levels."""

    def _init_(self, name: str, device_id: str, brightness: int = 50):
        super()._init_(name, device_id)
        # Validation: Brightness must be between 0 and 100
        if not (0 <= brightness <= 100):
            raise ValueError("Brightness must be between 0 and 100.")
        self.__brightness = brightness

    @property
    def brightness(self) -> int:
        return self.__brightness

    @brightness.setter
    def brightness(self, value: int):
        if 0 <= value <= 100:
            self.__brightness = value
        else:
            print("Error: Brightness must be between 0 and 100.")

    def increase_brightness(self, amount: int = 10):
        if not self.power_status:
            print(f"Cannot adjust brightness. [{self.name}] is OFF.")
            return
        self.brightness += amount
        print(f"[{self.name}] brightness increased to {self.__brightness}%.")

    def decrease_brightness(self, amount: int = 10):
        if not self.power_status:
            print(f"Cannot adjust brightness. [{self.name}] is OFF.")
            return
        self.brightness -= amount
        print(f"[{self.name}] brightness decreased to {self.__brightness}%.")


# --- 4. CHILD CLASS: SecurityCamera ---
class SecurityCamera(SmartDevice):
    """Child class inheriting from SmartDevice with recording functionalities."""

    def _init_(self, name: str, device_id: str):
        super()._init_(name, device_id)
        self.__recording_status = False

    @property
    def recording_status(self) -> bool:
        return self.__recording_status

    def start_recording(self):
        if not self.power_status:
            print(f"Cannot record. [{self.name}] is OFF.")
            return
        self.__recording_status = True
        print(f"[{self.name}] has STARTED recording.")

    def stop_recording(self):
        self.__recording_status = False
        print(f"[{self.name}] has STOPPED recording.")


# --- 5. MENU-DRIVEN INTERFACE ---
def main():
    # Instantiate objects
    sensor = TemperatureSensor("Living Room Sensor", "TEMP-01", 24.5)
    light = SmartLight("Bedroom Light", "LGT-02", 70)
    camera = SecurityCamera("Front Door Camera", "CAM-03")

    devices = [sensor, light, camera]

    while True:
        print("\n" + "="*35)
        print(" SMART DEVICE SYSTEM MENU ")
        print("="*35)
        print("1. Display Device Information")
        print("2. Turn Device On")
        print("3. Turn Device Off")
        print("4. Read Temperature")
        print("5. Adjust Brightness")
        print("6. Start/Stop Recording")
        print("7. Exit")

        choice = input("Enter choice (1-7): ").strip()

        if choice == "1":
            print("\n--- Current Devices Info ---")
            for dev in devices:
                dev.display_info()

        elif choice == "2":
            print("\nSelect Device to Turn ON:")
            for i, dev in enumerate(devices, 1):
                print(f"{i}. {dev.name}")
            ch = int(input("Choice: ")) - 1
            if 0 <= ch < len(devices):
                devices[ch].turn_on()

        elif choice == "3":
            print("\nSelect Device to Turn OFF:")
            for i, dev in enumerate(devices, 1):
                print(f"{i}. {dev.name}")
            ch = int(input("Choice: ")) - 1
            if 0 <= ch < len(devices):
                devices[ch].turn_off()

        elif choice == "4":
            sensor.read_temperature()

        elif choice == "5":
            print("\n1. Increase Brightness\n2. Decrease Brightness")
            b_choice = input("Choice: ")
            amount = int(input("Enter amount (e.g., 10): "))
            if b_choice == "1":
                light.increase_brightness(amount)
            elif b_choice == "2":
                light.decrease_brightness(amount)

        elif choice == "6":
            print("\n1. Start Recording\n2. Stop Recording")
            rec_choice = input("Choice: ")
            if rec_choice == "1":
                camera.start_recording()
            elif rec_choice == "2":
                camera.stop_recording()

        elif choice == "7":
            print("Exiting system. Goodbye!")
            break
        else:
            print("Invalid choice! Try again.")

if _name_ == "_main_":
    main()
```
